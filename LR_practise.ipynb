```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
# 1. Simple Linear Regression: Study Hours â Exam Score
df = pd.read_csv("study_hours_score.csv")

X = df[['hours']]
y = df['score']

model = LinearRegression()
model.fit(X, y)

print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)

y_pred = model.predict(X)
print("RÂ² Score:", r2_score(y, y_pred))

pred = model.predict(pd.DataFrame({'hours': [9]}))
print("Predicted Exam Score (for 9 hours):", pred[0])

plt.scatter(X, y, color='lightblue', label='Actual Scores')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.scatter(9, pred[0], color='orange', s=100, label='Prediction for 9 hours')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Simple Linear Regression: House Size â Price
df = pd.read_csv("house_size_price.csv")

X = df[['size_m2']]
y = df['price']

model = LinearRegression()
model.fit(X, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)

y_pred = model.predict(X)
print("RÂ² Score:", r2_score(y, y_pred))

pred = model.predict(pd.DataFrame({'size_m2': [120]}))
print("Predicted Price for 120 mÂ²:", pred[0])

plt.scatter(X, y, color='lightpink', label='Actual Prices')
plt.plot(X, y_pred, color='blue', label='Regression Line')
plt.scatter(120, pred[0], color='green', s=100, label='Prediction for 120 mÂ²')
plt.xlabel('House Size (mÂ²)')
plt.ylabel('Price')
plt.title('House Size vs Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 3. Simple Linear Regression: Salary vs Experience

df = pd.read_csv("salary_experience.csv")

X = df[['experience_years']]
y = df['salary_k']

model = LinearRegression()
model.fit(X, y)

print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)

y_pred = model.predict(X)
print("RÂ² Score:", r2_score(y, y_pred))

pred = model.predict(pd.DataFrame({'experience_years': [7]}))
print("Predicted Salary (7 years):", pred[0])

plt.scatter(X, y, color='lightgreen', label='Actual Salary')
plt.plot(X, y_pred, color='orange', label='Regression Line')
plt.scatter(7, pred[0], color='red', s=100, label='Prediction for 7 years')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary (k)')
plt.title('Experience vs Salary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 4. Multiple Linear Regression: House Price ~ Size + Bedrooms

df = pd.read_csv("house_price_multi.csv")

X = df[['size_m2', 'bedrooms']]
y = df['price_k']

model = LinearRegression()
model.fit(X, y)

print("Coefficients (size, bedrooms):", model.coef_)
print("Intercept:", model.intercept_)

y_pred = model.predict(X)
print("RÂ² Score:", r2_score(y, y_pred))

pred = model.predict(pd.DataFrame({'size_m2': [85], 'bedrooms': [3]}))
print("Predicted Price (size=85, bedrooms=3):", pred[0])

plt.scatter(y, y_pred, color='purple', label='Predicted vs Actual')
plt.scatter(pred[0], pred[0], color='green', s=100, marker='o', label='Predicted House (85mÂ², 3 bedrooms)')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linewidth=2, label='Perfect Prediction')
plt.xlabel("Actual Price (k)")
plt.ylabel("Predicted Price (k)")
plt.title("Predicted vs Actual House Prices with Prediction Point")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5. Multiple Features: Advertising Spend â Sales

df = pd.read_csv("advertising_sales.csv")

X = df[['TV_spend', 'radio_spend', 'social_spend']]
y = df['sales_k']

model = LinearRegression()
model.fit(X, y)

print("Coefficients (TV, Radio, Social):", model.coef_)
print("Intercept:", model.intercept_)

y_pred = model.predict(X)
print("RÂ² Score:", r2_score(y, y_pred))

pred = model.predict(pd.DataFrame({'TV_spend': [160], 'radio_spend': [32], 'social_spend': [18]}))
print("Predicted Sales (k):", pred[0])

plt.scatter(y, model.predict(X), color='yellow', label='Predicted vs Actual')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', linewidth=2, label='Perfect Prediction')
plt.scatter(pred[0], pred[0], color='green', s=100, marker='o', label='Predicted Sales')
plt.xlabel("Actual Sales (k)")
plt.ylabel("Predicted Sales (k)")
plt.title("Predicted vs Actual Sales with Prediction Point")
plt.legend()
plt.grid(True)
plt.show()
```
